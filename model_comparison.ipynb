# MODEL COMPARISON


```python
tfidf_model_title={'Title_Accuracy': '0.91',
 'Title_Recall': '0.96','loss':'0.51'}
```

```python
tfidf_model_content={'Content_Accuracy': '0.931107',
 'Content_Recall': '0.948491',
 'loss': '0.48'}
```

```python
sequence_model_title={'Title_Accuracy': '0.9078327',
 'Title_Recall': '0.9280112',
 'loss': '.34'}
```

```python
sequence_model_content={'Content_Accuracy': '0.8305411',
 'Content_Recall': '0.8603366',
 'loss': '0.87'}
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=[10,5])
x_values = ['Model 1', 'Model 2', 'Model 3', 'Model 4']
y_values = [
            float(tfidf_model_title.get('Title_Accuracy'))*100, 
            float(tfidf_model_content.get('Content_Accuracy'))*100, 
            float(sequence_model_title.get('Title_Accuracy'))*100, 
            float(sequence_model_content.get('Content_Accuracy'))*100
           ]
plt.bar(x_values, y_values, color=['r','g','b','m'])
plt.xlabel('Keras Neural Network Model')
plt.ylabel('Performance (%)')
plt.ylim(70,100)
plt.title('Accuracy Comparison')
```

```python
fig, ax = plt.subplots(figsize=[10,5])

x_values = ['Model 1', 'Model 2', 'Model 3', 'Model 4']
y_values = [
            float(tfidf_model_title.get('Title_Recall'))*100, 
            float(tfidf_model_content.get('Content_Recall'))*100, 
            float(sequence_model_title.get('Title_Recall'))*100, 
            float(sequence_model_content.get('Content_Recall'))*100
           ]

plt.bar(x_values, y_values, color=['r','g','b','y'])
plt.xlabel('Keras Neural Network Model')
plt.ylabel('Performance (%)')
plt.ylim(80,100)
plt.title('Recall Comparison')
```

```python
fig, ax = plt.subplots(figsize=[5,5])

x_values = ['Model 1', 'Model 2', 'Model 3', 'Model 4']
y_values = [
            float(tfidf_model_title.get('loss')), 
            float(tfidf_model_content.get('loss')), 
            float(sequence_model_title.get('loss')), 
            float(sequence_model_content.get('loss'))
           ]

plt.bar(x_values, y_values, color=['r','g','b','y'])
plt.xlabel('Keras Neural Network Model')
plt.ylabel('loss ')

plt.title('loss Comparison')
```
